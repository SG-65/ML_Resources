```python
#pandas series
import pandas as pd
a=['ASIA','AMERICA','AFRICA','EUROPE','ANTERTICA']
b=[10,20,30,40,50]
print(pd.Series(a))
print()
print(pd.Series(b))
```

```python
#custom index
sub=['maths','science','social science','english','bengali']
marks=[80,75,60,85,90]
pd.Series(marks,index=sub)
```

```python
#series using dictionary
marks={
     'maths':80,
     'science':75,
     'social science':60,
     'english':85,
     'bengali':90
}
marksheet=pd.Series(marks)
print(marksheet)
```

```python
#size of series
print(marksheet.size)
```

```python
#datatype of series
print(marksheet.dtype)
```

```python
#is_unique in series
print(marksheet.is_unique)
```

```python
#index of series
print(marksheet.index)
```
