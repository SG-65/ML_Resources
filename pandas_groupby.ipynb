```python
#agg method
#pasing dict
#using pparticular function on specific column
genres.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
#passing list
genres.agg(['min','max'])
```

```python
#combination both syntax
genres.agg({
    'Runtime':['mean','max'],
    'IMDB_Rating':'mean',
    'Gross':['sum','min'],
    'Metascore':'min'
})
```

```python
#looping on groups
for group,data in genres:
    print(data)
```

```python
#find the highest rated movie of each genre
for group,data in genres:
     print((data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]))
```

```python
#apply()
#find nimber of movies starting with A for each 
def find_A(group):
    return group['Series_Title'].str.startswith('A').sum()
genres.apply(find_A)
```

```python
#find ranking of each movies in the group according to imdb score
def rank_movie(group):
    group['movie_rank']=group['IMDB_Rating'].rank(ascending='False')
    return group
genres.apply(rank_movie)
```

```python
#find normalized imdb rating group wise
def rank_normalize(group):
    group['nomalized_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
genres.apply(rank_normalize)
```

```python
ipl=pd.read_csv('C:/Users/spand/OneDrive/Desktop/jupyter notebook/.ipynb_checkpoints/.ipynb_checkpoints/datasets/deliveries.csv')
ipl.head()
```

```python
#find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
#find the batsman with max no.of sixes
sixes=ipl[ipl['batsman_runs']== 6]
sixes
sixes.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
#find batsman with most no,=. of 4s and 6s in last 5 overs
temp_df=ipl[ipl['over'] >= 16]
temp_df=temp_df[(temp_df['batsman_runs']==4) |(temp_df['batsman_runs']==6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
#find kohli record against all team
temp=ipl[ipl['batsman']=='V Kohli']
temp.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
#create a function that can return the highest score of any batsmans
def highest(batsman):
  temp=ipl[ipl['batsman']==batsman]
  return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)

highest('V Kohli')
```
